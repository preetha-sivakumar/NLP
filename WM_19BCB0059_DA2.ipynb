```python
import requests
URL ="https://en.wikipedia.org/wiki/BTS"
r = requests.get(URL)
print(r.content)
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(r.content, 'html5lib')
print(soup.prettify())
```

```python
# title
html_content = requests.get(URL).text
soup = BeautifulSoup(html_content,"html.parser")

print(soup.find('title').text)
```

```python
#p tag
texts = soup.find_all('p')
for text in texts:
    print(text.get_text())
```

```python
#a tag
texts = soup.find_all('a')
for text in texts:
    print(text.get_text())
```

```python
#div tag
texts = soup.find_all('div')
for text in texts:
    print(text.get_text())
```

```python
#textual content
print(soup.get_text())
```
